# **Historia:**
Un grupo de agricultores está participando en un concurso para ver quién tiene las zanahorias más grandes. Han medido el peso (en gramos) de 12 zanahorias y necesitan analizar los datos para ver qué tan uniformes son sus cultivos.
> Añadir blockquote



**Tu misión
 Escribe un programa en Python que haga lo siguiente:**
1. Calcule la media (promedio) del peso de las zanahorias.
2. Encuentre la mediana del peso.
3. Determine la moda (peso más común).
4. Calcule la desviación estándar para ver qué tanto varían los pesos.
5. Muestre los resultados de forma clara y divertida.

**Estos son los pesos (en gramos) de las zanahorias concursantes:**
zanahorias = 1300, 280, 290, 310, 275, 290, 295, 315, 290, 280, 310, 305]
**Bonus (Opcional)**
ll Extra 1: Agrega una visualización con Matplotlib & para mostrar la distribución de los pesos. el Extra 2: Modifica el código para que el usuario pueda ingresar los pesos de nuevas zanahorias.

```python
import numpy as np
import statistics as stats
import matplotlib.pyplot as plt

# Pesos de las zanahorias en gramos (actualizado según la descripción)
zanahorias = [300, 280, 290, 310, 275, 290, 295, 315, 290, 280, 310, 305]

# Estadísticas descriptivas
media = np.mean(zanahorias) # Promedio
mediana = np.median(zanahorias) # Valor central
moda = stats.mode(zanahorias) # Valor más común
desviacion = np.std(zanahorias) # Variabilidad

print(f"Media: {media:.2f}g")
print(f"Mediana: {mediana:.2f}g")
print(f"Moda: {moda}g")
print(f"Desviación estándar: {desviacion:.2f}g")
```

### **Visualización de la Distribución de Pesos (Bonus Extra 1)**

Vamos a crear un histograma para ver cómo se distribuyen los pesos de las zanahorias. Esto nos permitirá identificar rápidamente si los pesos están concentrados en un rango específico o si hay zanahorias con pesos muy diferentes al resto.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(zanahorias, bins=5, edgecolor='black', alpha=0.7)
plt.title('Distribución de Pesos de las Zanahorias')
plt.xlabel('Peso (gramos)')
plt.ylabel('Frecuencia')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

### **Ingreso de Pesos por el Usuario (Bonus Extra 2)**

Aquí puedes ingresar los pesos de nuevas zanahorias. Deberás separarlos por comas. Después de ingresar los nuevos pesos y ejecutar esta celda, puedes volver a ejecutar las celdas de estadísticas y visualización para ver los resultados actualizados.

```python
# Permite al usuario ingresar nuevos pesos de zanahorias
pesos_str = input("Ingresa los pesos de las zanahorias separados por comas (ejemplo: 280, 290, 310): ")

# Crea una lista vacía para almacenar los nuevos pesos
nueva_zanahorias = []

# Divide la cadena de texto por comas y convierte cada parte a un número entero
for peso_str in pesos_str.split(','):
    nueva_zanahorias.append(int(peso_str.strip()))

# Actualiza la lista original de zanahorias
zanahorias = nueva_zanahorias
print(f"Lista de zanahorias actualizada: {zanahorias}")

# Puedes volver a ejecutar las celdas de estadísticas (xDiqwOCaPgYZ) y visualización (86e955fc)
# para ver los resultados con esta nueva lista.
```
