```python
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
# Si usas stemming para ingles, también necesitarías:
#from nltk.stem import PorterStemmer
from nltk.stem.snowball import SnowballStemmer # para español
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')


```

```python
frase_original = "¡Los científicos han descubierto que el 🌞 (sol) emite partículas energéticas a una velocidad de 1.5 millones de kilómetros por hora!"

print(f"Frase original: {frase_original}\n")

# 1. Normalización a minúsculas
frase_minusculas = frase_original.lower()
print(f"1. Después de normalizar a minúsculas: {frase_minusculas}\n")

# 2. Eliminación de puntuación y emojis
# Usamos expresiones regulares para quitar todo lo que no sea una letra o un espacio
frase_sin_puntuacion = re.sub(r'[^\w\s]', '', frase_minusculas)
print(f"2. Después de eliminar puntuación y emojis: {frase_sin_puntuacion}\n")

# 3. Eliminación de números
frase_sin_numeros = re.sub(r"\d+", "", frase_sin_puntuacion)
print(f"3 Después de eliminar números: {frase_sin_numeros}\n")

#4. Eliminamos las stopwords
stop_words_es = set(stopwords.words('spanish'))
tokens = word_tokenize(frase_sin_numeros)
frase_sin_stopwords = [word for word in tokens if word not in stop_words_es]

print(f"4. Después de eliminar stopwords: {frase_sin_stopwords}\n")

#5. Lematización

lematizer = WordNetLemmatizer()
frase_lemmatizada = [lematizer.lemmatize(word) for word in frase_sin_stopwords]
print(f"5. Después de lematización: {frase_lemmatizada}\n")

# 6. Steming
stemmer_es = SnowballStemmer("spanish")
frase_stemmed = [stemmer_es.stem(word) for word in frase_sin_stopwords]
print(f"6. Después de stemming (con SnowballStemmer para español): {frase_stemmed}")




```
