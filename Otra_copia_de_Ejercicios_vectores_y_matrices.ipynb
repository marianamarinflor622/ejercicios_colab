1. El Registro de Temperatura (Lista)
Tienes una lista de temperaturas de la semana. Suma una "corrección" de 2 grados a cada una.

Dato: temps = [18, 20, 22]

Reto: Crea una nueva lista donde cada valor sea v + 2.

```python
temperatura = [18, 20, 22]
nueva_temperatura = [v + 2 for v in temperatura]#Sumale 2 a v para cada v en la lista temperatura, buscar acerca de List Comprehension
print(nueva_temperatura)
```

3. El Carrito de Compras (Lista)
Tienes los precios de tres productos.

Dato: precios = [10.5, 5.0, 2.99]

Reto: Usa la función sum(precios) para obtener el total. Luego, usa len(precios) para saber cuántos productos hay.

```python
precios = [10.5, 5.0, 2.99]
total=sum(precios)
cantidad=len(precios)
print(f"La suma es: {total:.2f}")
print("Total:", round(total, 2))
print(f"La suma es: {round(total, 2)}")
print(f"La cantidad total de productos es: {cantidad}")
```

8. Suma de Filas (Lista de Listas)
Tienes una matriz donde cada fila es el gasto diario de una persona en 2 días.

Dato: gastos = [[10, 5], [20, 30]]

Reto: Usa un bucle for para imprimir la suma de cada fila.

```python
gastos = [[10, 5], [20, 30]]
for persona in gastos:
    print(sum(persona))
```

9. Transposición Básica (Lista de Tuplas)
A veces los datos vienen como una lista de filas, donde cada fila es una tupla.

Dato: data = [(1, 2), (3, 4)]

Reto: Crea una nueva lista que contenga solo el primer elemento de cada tupla: [1, 3].

```python
data = [(1, 2), (3, 4)]
lista=[]
for tupla in data:
    lista.append(tupla[0])
print(lista)
```

10. El Filtro de Imagen (Lista de Listas)
Representa una "imagen" de 2x2 píxeles (escala de grises).

Dato: pixeles = [[255, 0], [128, 64]]

Reto: Multiplica cada valor por 0.5 para "oscurecer" la imagen a la mitad.

```python
pixeles = [[255, 0], [128, 64]]
for fila in pixeles:
    for pixel in fila:
        print(int(pixel*0.5))
```
