```python
import pandas as pd
import matplotlib.pyplot as plt

datos = {
    'Día': ['Lun', 'Mar', 'Mie', 'Jue', 'Vie', 'Sab', 'Dom'],
    'Temp': [22, 24, 23, 28, 26, 30, 31]
}

df=pd.DataFrame(datos)
#print(df)

#empezamos a graficar
plt.plot(df["Día"], df['Temp'], marker="o",color="pink" )
plt.title("evolucion de la temperatura")
plt.xlabel("Días de la semana")
plt.ylabel("grados celsius")
plt.show()
```

La Evolución del Ahorro: Crea un gráfico de líneas que muestre tus ahorros durante 6 meses. (Eje X: Meses, Eje Y: Dinero).

```python
from matplotlib.lines import lineStyles
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun']
ahorros = [100, 150, 130, 200, 250, 300]

plt.plot(meses,ahorros, marker="s", linestyle="--",color='pink')
plt.title('Evolución de mis ahorros')
plt.xlabel('Meses')
plt.ylabel('Euros €')
plt.show()
```

Competencia de Frutas: Crea un gráfico de barras comparando la cantidad de inventario de 4 frutas: Manzanas (50), Bananas (30), Cerezas (45) y Peras (20).



```python

frutas = ['Manzanas', 'Bananas', 'Cerezas', 'Peras']
cantidades = [50, 30, 45, 20]
plt.bar(frutas, cantidades, color=['#ED0909', 'yellow', '#BA0D0D', 'lightgreen'])
plt.title("Inventario de frutas")
plt.xlabel("Frutas")
plt.ylabel("Cantidad")
plt.show()
```

***Relacion Estatura-Peso*** :
usa un grafico de dispersión con (scatter) con 5 puntos de datos inventados para ver una operer si hay una relacion visual entre la altura de una persona y su peso.

```python
estaturas = [160, 170, 180, 155, 190]
pesos = [60, 72, 85, 50, 95]
plt.scatter(estaturas, pesos, color='purple', s=100, marker='o')
plt.title("Relación Estatura-Peso")
plt.xlabel("Estatura (cm)")
plt.ylabel("Peso (kg)")
plt.grid(True)
plt.show()
```

 Imagina que quieres analizar cómo gastas tu tiempo un día de 24 horas. Este es un caso de uso clasico para un grafico de sectores.

```python
actividades = ['Dormir', 'Comer', 'Trabajo/Estudio', 'Libre', 'Deporte']
horas = [8, 2, 8, 4, 2]
desfase=[0, 0, 0, 0.1, 0]
colores = ['skyblue', 'orange', 'lightgreen', 'pink', 'gold']
plt.pie(horas,labels=actividades, colors=colores, explode=desfase,startangle=140, autopct="%1.2f%%" )
plt.title("Tiempo en un día")
#plt.legend(actividades, loc="upper left", bbox_to_anchor=(1, 0.5))
plt.show()
```

Distribución de Exámenes: Imagina que tienes una lista de 20 notas de alumnos (del 1 al 10). Crea un histograma para ver qué notas son las más frecuentes.

```python
notas = [1, 5, 8, 9, 7, 10, 4, 6, 8, 8, 9, 5, 3, 7, 8, 10, 2, 9, 8, 7]
#plt.hist(notas, bins=5, color='orange',edgecolor="black")
plt.hist(
    notas,
    bins=range(1, 11),
    color='mediumpurple',
    edgecolor='black',
    rwidth=1
)
plt.title("Distribución de Exámenes")
plt.xlabel("Nota 1-10")
plt.ylabel("Numero de alumnos/as")
plt.show()
```
