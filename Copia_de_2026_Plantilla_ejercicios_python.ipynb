Enunciado: Crea una variable energia = 5. Mientras la energía sea mayor a 0, imprime "Corriendo..." y resta 1 a la energía. Al final imprime "¡Me cansé!".

**Juego de adivinanza de numeros ** el programa generaro un nümero aleatorio del 1, y 100,y el usuario tendrá que adivinarlo. El programa dara pistas si el número es mayor o menor hasta que el usuario acierte.

```python
import random

def juego_de_adivinanza():
    numero_secreto = random.randint(1, 100)
    intentos = 0
    adivinado = False
    max_intentos = 5  # límite de intentos

    print("¡Bienvenida a mi juego de adivinanza de números!")
    print("Estoy pensando en un número entre 1 y 100.")
    print(f"Tienes {max_intentos} intentos. ¡Trata de adivinarlo!")

    while not adivinado and intentos < max_intentos:
        intentos_restantes = max_intentos - intentos
        print(f"\nIntentos restantes: {intentos_restantes}")
        numero_introducido = int(input("Adivina el número: "))
        intentos += 1

        if numero_introducido < numero_secreto:
            print("El número es mayor. ¡Sigue intentando!")
        elif numero_introducido > numero_secreto:
            print("El número es menor. ¡Sigue intentando!")
        else:
            print(f"¡Felicidades! Adivinaste el número {numero_secreto} en {intentos} intentos.")
            adivinado = True

    if not adivinado:
        print(f"\n¡Se acabaron los intentos! El número secreto era {numero_secreto}.")

juego_de_adivinanza()
```

```python
META_AGUA  = 3.0
META_PASOS = 10000
META_SUENO = 8

nombre = input("¿Cómo te llamas? ")
print(f"\n¡Hola, {nombre}! Vamos a revisar tu salud de hoy 💪\n")

agua  = float(input("💧 ¿Cuántos litros de agua bebiste hoy? "))
pasos = int(input("👟 ¿Cuántos pasos diste hoy? "))
sueno = float(input("😴 ¿Cuántas horas dormiste anoche? "))

print(f"\n======= 📊 RESUMEN DE {nombre.upper()} =======")
print(f"💧 Agua:  {agua}L de {META_AGUA}L")
print(f"👟 Pasos: {pasos} de {META_PASOS}")
print(f"😴 Sueño: {sueno}h de {META_SUENO}h")
print("==========================================")
```

```python

def analizador_salud():
    print("🏃 Bienvenido al Analizador de Salud")
    nombre = input("¿Cómo te llamas? ")
    print(f"\nHola {nombre}, vamos a analizar tu salud 💪\n")

    # Recomendaciones fijas
    agua_recomendada  = 3      # litros
    sueño_recomendado = 8      # horas
    pasos_recomendados = 10000

    # Agua
    agua = float(input("💧 ¿Cuántos litros de agua has bebido hoy? "))
    if agua >= agua_recomendada:
        print("✅ Bien, estás bien hidratado/a.")
    else:
        print(f"❌ Te faltan {agua_recomendada - agua} litros para la recomendación.")

    # Sueño
    sueño = float(input("😴 ¿Cuántas horas has dormido? "))
    if sueño >= sueño_recomendado:
        print("✅ Buen descanso.")
    else:
        print(f"❌ Te faltan {sueño_recomendado - sueño} horas de sueño.")

    # Pasos
    pasos = int(input("👟 ¿Cuántos pasos has hecho hoy? "))
    if pasos >= pasos_recomendados:
        print("✅ Excelente actividad física.")
    else:
        print(f"❌ Te faltan {pasos_recomendados - pasos} pasos.")

    print("\n🎯 Análisis completado. ¡Cuida tu salud!")

# Ejecutar el programa
analizador_salud()
```
