Conjunción (∧) - "Y" → Ambas deben ser verdaderas para que toda la oración sea verdadera.

```python
calor = True

negacion = not calor

print(f"proposicion original,Hace Calor:{calor}")
print(f"Negacion de la proposición original , hace calor:{negacion}")
```

```python
hoy_llueve = True
hace_frio= False
frase = hoy_llueve and hace_frio
print(frase)

#para que tenga que ser verdadero ambas tiene que ser verdaderas.
```

Disyunción (∨) - "O" → Basta que una sea verdadera para que todo sea verdadero.

```python
voy_al_cine=True
voy_a_comer=False
frase = voy_al_cine or voy_a_comer
print(frase)
```

```python
sabe_python = True
sabe_mates = False
experiencia_anios = 3

entra = sabe_python and (sabe_mates or experiencia_anios >=5 )
print(entra)
```

**ejercicio para verificar cuanta hambre tienes con if elif y else **

```python
hambre = input ("Cuanta hambre tienes? Elige una opción : mucha/ mas_o_menos/poca")
if hambre == "mucha":
  print ("Come una pizza grande!")
elif hambre == "mas_o_menos"
  print ("come una manzana.")
elif hambre == "poca"
  print ("Toma un vaso de agua")
else
  print("vete a correr")
```
